```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("bengaluru_house_prices.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(df.price,bins=500,rwidth=0.6)
plt.xlabel("Price")
plt.ylabel("count")
plt.yscale("log")
plt.show()
```

```python
maxthreshold,minthreshold=(df.price.quantile(0.999),df.price.quantile(0.001))
maxthreshold,minthreshold
```

```python
df[df.price<minthreshold]
```

```python
df[df.price>maxthreshold]
```

```python
df2=df[(df.price>minthreshold)&(df.price<maxthreshold)]
```

```python
df2.shape
```

```python
plt.hist(df2.price,bins=30,rwidth=0.8)
plt.xlabel("Price")
plt.ylabel("count")
plt.show()
```

```python
df.shape
```

```python
df2.describe()
```

```python
up=df2.price.mean()+3*df.price.std()
ll=df2.price.mean()-3*df.price.std()
up,ll
```

```python
df3=df2[(df2.price>ll)&(df2.price<up)]
```

```python
df3.shape
```

```python
plt.hist(df3.price,bins=30,rwidth=0.8)
plt.xlabel("Price")
plt.ylabel("count")
plt.show()
```

```python
from scipy.stats import norm

plt.hist(df3.price, bins=20, rwidth=0.8, density=True)
plt.xlabel('Price')
plt.ylabel('Count')

rng = np.arange(-200, df3.price.max(), 10)
plt.plot(rng, norm.pdf(rng,df3.price.mean(),df3.price.std()))
```

```python
df2["zscore"]=(df.price-df.price.mean())/df.price.std()
```

```python
df2
```

```python
df4=df2[(df2.zscore<3)&(df2.zscore>-3)]
```
